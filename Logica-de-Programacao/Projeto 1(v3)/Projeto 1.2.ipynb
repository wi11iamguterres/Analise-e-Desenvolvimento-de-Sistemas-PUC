```python
#Inicialização dos contadores
cont_escaldante=0
mes_mais_quente=0
mes_menos_quente=0

#Lista de meses
meses=["Janeiro",
       "Fevereiro",
       "Março",
       "Abril",
       "Maio",
       "Junho",
       "Julho",
       "Agosto",
       "Setembro",
       "Outubro",
       "Novembro",
       "Dezembro"]

#Inicialização da lista de temperaturas
temperaturas=[]

#Inserção da temperatura nos meses
for m in range(1,13):
  while True:
    try:
      temp=float(input("Informe a temperatura entre -60°C a +50C para o mês {}: ".format(m)))
      if temp >-60 and temp <50:
        temperaturas.append(temp)
        break
      else:
        print("Temperatura inválida!")
    except ValueError:
      print("Caractere inválido!")

#Soma das temperaturas
soma_temp=sum(temperaturas)

#Cáculo da média anual de temperatura
media_anual=soma_temp/len(temperaturas)
print("A média anual é: {:.2f}°C.".format(media_anual))

#Cálculo da quantidade de meses com temperatura acima dos 33°C
for temp in temperaturas:
  if temp>33:
    cont_escaldante=cont_escaldante+1
print("A quantidade de meses com temperaturas acima de 33°C é: {}".format(cont_escaldante))

#Apresentação do mês mais quente
mes_mais_quente=temperaturas.index(max(temperaturas))
print("O mês mais quente é: {}".format(meses[mes_mais_quente]))

#Apresentação do mês mais frio
mes_menos_quente=temperaturas.index(min(temperaturas))
print("O mês menos quente é: {}".format(meses[mes_menos_quente]))
```
