```python
def lista_temp():
  lista=[]
  for i in range(1,4):
    while True:
      try:
        temperatura=float (input("Digite a temperatura do mês {}: ".format(i)))
        if temperatura < -60 or temperatura > 50:
          print("Temperatura inválida! Digite uma temperatura entre -60°C a +50°C.")
          temperatura=float(input("Redigite a temperatura do mês {}: ".format(i)))
          lista.append(temperatura)
          break
        else:
          lista.append(temperatura)
          break
      except ValueError:
        print("Valor inválido! Digite um número entre -60°C a +50°C.")
  return lista
print(lista_temp())

def media_anual(lista):
  media=sum(lista)/len(lista)
  return media
  print("A média anual de temperatura é de: {}°C.".format(media))

print(media_anual())


3
```
